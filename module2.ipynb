```python
import pandas as pd
import numpy as np
conn = sqlite3.connect(':memory:')
conn.executescript('''
    CREATE TABLE Customers (customer_id INT, customer_name TEXT);
    CREATE TABLE Products (product_id INT, product_name TEXT);
    CREATE TABLE Orders (order_id INT, customer_id INT, product_id INT);
    INSERT INTO Customers VALUES (1, 'Shubh Mittal'), (2, 'Bob Jones');
    INSERT INTO Products VALUES (101, 'MacBook Air'), (102, 'Sony Headphones');
    INSERT INTO Orders VALUES (7001, 1, 101), (7002, 2, 102);
''')
query = """
SELECT
    o.order_id,
    c.customer_name,
    p.product_name
FROM Orders o
JOIN Customers c ON o.customer_id = c.customer_id
JOIN Products p ON o.product_id = p.product_id
"""
df = pd.read_sql_query(query, conn)
print(df.to_string(index=False))
```

```python
import pandas as pd
import numpy as np
conn = sqlite3.connect(':memory:')
conn.execute('''
    CREATE TABLE Employees (
        employee_id INT,
        employee_name TEXT,
        manager_id INT
    )
''')
data = [
    (1, 'Shubh Mittal', None),  # The CEO
    (2, 'Michael Scott', 1),  # Reports to Sarah
    (3, 'Dwight Schrute', 2), # Reports to Michael
    (4, 'Jim Halpert', 2)     # Reports to Michael
]
conn.executemany('INSERT INTO Employees VALUES (?,?,?)', data)
query = """
SELECT
    e.employee_name AS Employee,
    COALESCE(m.employee_name, 'Top Level') AS Manager
FROM Employees e
LEFT JOIN Employees m ON e.manager_id = m.employee_id
"""
df = pd.read_sql_query(query, conn)
print(df.to_string(index=False))
```

```python
import pandas as pd
import numpy as np
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
cursor.execute('''
    CREATE TABLE Employees (
        emp_id INTEGER PRIMARY KEY,
        name TEXT,
        salary INTEGER
    )
''')
initial_employees = [
    (101, 'Shubh Mittal', 50000),
    (102, 'Bob', 60000)
]
cursor.executemany('INSERT INTO Employees (emp_id, name, salary) VALUES (?, ?, ?)', initial_employees)
conn.commit()
print("Initial Employees:")
df_initial = pd.read_sql_query("SELECT * FROM Employees", conn)
print(df_initial.to_string(index=False))
new_employee_data = [
    (101, 'Shubh Mittal', 55000),
    (103, 'Charlie', 70000)
]
for emp_id, name, salary in new_employee_data:
    cursor.execute('''
        INSERT INTO Employees (emp_id, name, salary) VALUES (?, ?, ?)
        ON CONFLICT(emp_id) DO UPDATE SET
            name = excluded.name,
            salary = excluded.salary
    ''', (emp_id, name, salary))
conn.commit()
print("\nEmployees after MERGE (UPSERT):")
df_final = pd.read_sql_query("SELECT * FROM Employees", conn)
print(df_final.to_string(index=False))
conn.close()
```
