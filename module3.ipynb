```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'months': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'sales': [100, 120, 130, 110, 140, 150],
    'revenue': [1000, 1200, 1350, 1100, 1450, 1500]
}
df_sales = pd.DataFrame(data)
display(df_sales)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
# Create a figure and an axes object
plt.figure(figsize=(10, 6))
# Plot sales
sns.lineplot(x='months', y='sales', data=df_sales, marker='o', label='Sales')
# Plot revenue
sns.lineplot(x='months', y='revenue', data=df_sales, marker='x', label='Revenue')
plt.title('Sales and Revenue Over Months')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# data fetch using plotly express
#upload data set
px.data.gapminder()
```

```python
# data fetch using plotly express
#upload data set
canada_data=px.data.gapminder().query("country=='Canada'")
```

```python
#bar chart using plotly
fig=px.bar(canada_data,x="year",y="pop")
fig.show()
```

```python
df=px.data.gapminder()
df
```

```python
# Adding title
fig=px.bar(canada_data,x='year',y='pop',title="population growth of canada over a year")
fig.show()
```

```python
# Adding labels
fig=px.bar(canada_data,x='year',y='pop',title="population growth of canada over a year",labels={
     'year':'years',
     'pop':'population'
     })
fig.show()
```

```python
# Adding color
fig=px.bar(canada_data,x='year',y='pop',title="population growth of canada over a year",labels={
     'year':'years',
     'pop':'population'
     },color="lifeExp")
fig.show()
```

```python
# Adding color
fig=px.bar(canada_data,x='year',y='pop',title="population growth of canada over a year",labels={
     'year':'years',
     'pop':'population'
     },color="pop")
fig.show()
```

```python
# Adding more hover data
fig=px.bar(canada_data,x='year',y='pop',title="population growth of canada over a year",labels={
     'year':'year',
     'pop':'population'
     },color="pop",hover_data=['lifeExp','gdpPercap'])
fig.show()
```

```python
# Adding value at top of the bar
fig=px.bar(canada_data,x='year',y='pop',title="population growth of canada over a year",labels={
     'year':'year',
     'pop':'population'
     },color="pop",hover_data=['lifeExp','gdpPercap'],text="pop")
fig.show()
```

```python
#update traces
fig=px.bar(canada_data,x='year',y='pop',title="population growth of canada over a year",labels={
     'year':'year',
     'pop':'population'
     },color="pop",hover_data=['lifeExp','gdpPercap'],text="pop")
fig.update_traces(texttemplate='%{text:.2s}',textposition='outside')
fig.show()
```

```python
# updateting layout
fig=px.bar(canada_data,x='year',y='pop',title="population growth of canada over a year",labels={
     'year':'year',
     'pop':'population'
     },color="pop",hover_data=['lifeExp','gdpPercap'],text="pop")
fig.update_traces(texttemplate='%{text:.2s}',textposition='outside')
fig.update_layout(xaxis_tickangle=-50)
fig.show()
```

```python
# updateting layout
fig=px.bar(canada_data,x='year',y='pop',title="population growth of canada over a year",labels={
     'year':'year',
     'pop':'population'
     },color="pop",hover_data=['lifeExp','gdpPercap'],text="pop")
fig.update_traces(texttemplate='%{text:.2s}',textposition='outside')
fig.update_layout(xaxis_tickangle=-50,yaxis_title="population of canada")
fig.show()
```
